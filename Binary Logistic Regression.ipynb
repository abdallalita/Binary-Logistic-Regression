Mr John would like you to create a Logistic Regression Model an Associated ROC curve for his cancer.csv dataset
in order to predict if the patient has cancer or not.
Dependent variable is Class:Benign=No cancer (2)
                            Malignant=cancer (4)


```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```python

data=pd.read_csv("cancer.csv")
data
```

```python
data.drop(["id"],axis="columns",inplace=True)
data
```

```python
data.isna().any()
```

```python
data.duplicated()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data
```

```python
data.dtypes
```

```python
data.shape
```

```python
data.info()
```

```python
data["Class"] = data["Class"].astype("category")
data.dtypes
```

```python
data["Class"] = data["Class"].cat.codes
```

```python
data
```

```python
y=data["Class"]
print(y)
```

```python
X=data[["Clump Thickness","UofCSize","UofCShape","Marginal Adhesion","SECSize","Bare Nuclei","Bland Chromatin","Normal Nucleoli","Mitoses"]]
print(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
model= LogisticRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.predict([[4,8,8,5,4,5,10,4,1]])
```

model.predict([[4,8,8,5,4,5,10,4,1]])

```python
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(model.score(X_test, y_test)))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
